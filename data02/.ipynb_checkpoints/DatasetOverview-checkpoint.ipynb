```python
import json
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

# Load the dataset

## Load the index
It contains the image paths, and the 6DOF poses in format (tx,ty,tz,qx,qy,qz,qw).

```python
with open("map.json","r") as f:
    m_idx=json.load(f)
    m_ims=np.array(m_idx["im_paths"])
    m_poses=np.array(m_idx["poses"])
with open("query.json","r") as f:
    q_idx=json.load(f)
    q_ims=np.array(q_idx["im_paths"])
    q_poses=np.array(q_idx["poses"])
```

```python
plt.scatter(m_poses[:,0],m_poses[:,1], label="Map")
plt.scatter(q_poses[:,0],q_poses[:,1], label="Query")
plt.legend()
```

## Load the ground truth
The gt.h5 file contains two datasets under the keys sim and fov:
1. sim: 100x100 binary matrix reflecting whether two images are similar (1) or not (0).
2. fov: 100x100 float matrix containing the degree of similarity between two images [0-1].

The matrices are accesed as matrix[query_idx, map_idx]

```python
with h5py.File("gt.h5","r") as f:
    fovs=f["fov"][:]
    sim=f["sim"][:]
```

```python
plt.subplot(1,2,1)
plt.imshow(sim)
plt.xlabel("Query")
plt.ylabel("Map")
plt.title("Similar pairs")

plt.subplot(1,2,2)
plt.imshow(fovs)
plt.xlabel("Query")
plt.ylabel("Map")
plt.title("Field of view overlap")
plt.show()
```

## Visualization of random pairs

```python
hard_positives=np.where(fovs>=0.75)
soft_positives=np.where(np.logical_and(fovs<0.75, fovs>=0.5))
soft_negatives=np.where(np.logical_and(fovs<0.5, fovs>0))
hard_negatives=np.where(fovs==0)
```

```python
hp_idx=np.random.randint(len(hard_positives[0]))
hp_map_idx=hard_positives[1][hp_idx]
hp_query_idx=hard_positives[0][hp_idx]
hp_fov=fovs[hp_query_idx,hp_map_idx]
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.title(str(hp_fov))
plt.imshow(plt.imread(m_ims[hp_map_idx]))
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(plt.imread(q_ims[hp_query_idx]))
plt.axis("off")
```

```python
sp_idx=np.random.randint(len(soft_positives[0]))
sp_map_idx=soft_positives[1][sp_idx]
sp_query_idx=soft_positives[0][sp_idx]
sp_fov=fovs[sp_query_idx,sp_map_idx]
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.title(str(sp_fov))
plt.imshow(plt.imread(m_ims[sp_map_idx]))
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(plt.imread(q_ims[sp_query_idx]))
plt.axis("off")
```

```python
sn_idx=np.random.randint(len(soft_negatives[0]))
sn_map_idx=soft_negatives[1][sn_idx]
sn_query_idx=soft_negatives[0][sn_idx]
sn_fov=fovs[sn_query_idx,sn_map_idx]
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.title(str(sn_fov))
plt.imshow(plt.imread(m_ims[sn_map_idx]))
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(plt.imread(q_ims[sn_query_idx]))
plt.axis("off")
```
